```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = plt.imread('mypic.jpg')
```

```python
plt.imshow(image)
```

```python
model = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
face  = model.detectMultiScale(image)
```

```python
for i in face:
    x,y,w,h = i
    image = cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), thickness=4, lineType=8)
```

```python
plt.imshow(image)
```
